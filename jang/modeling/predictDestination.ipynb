```python
import pandas as pd

base_path = r'C:\Users\hanaj\OneDrive\문서\[강남구혁신인재육성아카데미]데이터분석\미니프로젝트\국내여행로그데이터'
sub_place = ['동부권', '서부권', '수도권', '제주도 및 도서지역']
sub_folder = ['TS_csv', 'VS_csv']

# 전국 단위 데이터를 쌓기 위한 리스트
travel_merge = []
traveller_merge = []
companion_merge = []

for place in sub_place:
    for folder in sub_folder:
        travel_path = base_path + "\\" + place + "\\" + folder

        suffix = {
            '동부권': 'B',
            '서부권': 'C',
            '수도권': 'A',
            '제주도 및 도서지역': 'D'
        }[place]

        travel = pd.read_csv(travel_path + f"\\tn_travel_여행_{suffix}.csv")
        traveller = pd.read_csv(travel_path + f"\\tn_traveller_master_여행객 Master_{suffix}.csv")
        companion = pd.read_csv(travel_path + f"\\tn_companion_info_동반자정보_{suffix}.csv")
        
        travel_merge.append(travel)
        traveller_merge.append(traveller)
        companion_merge.append(companion)

# ==== 여기서부터 전국 단위 합치기 ====
travel_merge_all = pd.concat(travel_merge).reset_index(drop=True)
traveller_merge_all = pd.concat(traveller_merge).reset_index(drop=True)
companion_merge_all = pd.concat(companion_merge).reset_index(drop=True)

travel_merge_all['TRAVEL_END_YMD'] = pd.to_datetime(travel_merge_all['TRAVEL_END_YMD'])
travel_merge_all['TRAVEL_START_YMD'] = pd.to_datetime(travel_merge_all['TRAVEL_START_YMD'])
travel_merge_all['TRAVEL_DAYS'] = (travel_merge_all['TRAVEL_END_YMD'] - travel_merge_all['TRAVEL_START_YMD']).dt.days

TravelID = travel_merge_all[['TRAVEL_ID','TRAVELER_ID','TRAVEL_DAYS','MVMN_NM']]
traveller_sel = traveller_merge_all[['TRAVELER_ID','TRAVEL_TERM','TRAVEL_NUM','TRAVEL_STATUS_RESIDENCE','TRAVEL_STATUS_DESTINATION','TRAVEL_COMPANIONS_NUM', 'TRAVEL_STATUS_ACCOMPANY', 'INCOME', 'TRAVEL_STYL_1','TRAVEL_STYL_2','TRAVEL_STYL_3','TRAVEL_STYL_4','TRAVEL_STYL_5','TRAVEL_STYL_6','TRAVEL_STYL_7','TRAVEL_STYL_8', 'TRAVEL_MOTIVE_1','TRAVEL_MOTIVE_2','TRAVEL_MOTIVE_3','TRAVEL_LIKE_SIDO_1','TRAVEL_LIKE_SGG_1','TRAVEL_LIKE_SIDO_2','TRAVEL_LIKE_SGG_2','TRAVEL_LIKE_SIDO_3','TRAVEL_LIKE_SGG_3']].dropna()
companion_sel = companion_merge_all[['TRAVEL_ID', 'REL_CD','COMPANION_AGE_GRP','COMPANION_SITUATION']]

all_merge = pd.merge(TravelID, traveller_sel, on='TRAVELER_ID', how='left')
all_merge = pd.merge(all_merge, companion_sel, on='TRAVEL_ID', how='left')
all_merge = all_merge.dropna()

all_merge
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

y = all_merge['TRAVEL_STATUS_DESTINATION']
x = all_merge[['TRAVEL_DAYS','MVMN_NM','TRAVEL_TERM','TRAVEL_NUM','REL_CD','COMPANION_AGE_GRP','COMPANION_SITUATION','TRAVEL_STATUS_RESIDENCE','TRAVEL_COMPANIONS_NUM', 'TRAVEL_STATUS_ACCOMPANY', 'INCOME', 'TRAVEL_STYL_1','TRAVEL_STYL_2','TRAVEL_STYL_3','TRAVEL_STYL_4','TRAVEL_STYL_5','TRAVEL_STYL_6','TRAVEL_STYL_7','TRAVEL_STYL_8', 'TRAVEL_MOTIVE_1','TRAVEL_MOTIVE_2','TRAVEL_MOTIVE_3','TRAVEL_LIKE_SIDO_1','TRAVEL_LIKE_SGG_1','TRAVEL_LIKE_SIDO_2','TRAVEL_LIKE_SGG_2','TRAVEL_LIKE_SIDO_3','TRAVEL_LIKE_SGG_3']]

x_train_dummies = pd.get_dummies(x)


X_train, X_test, y_train, y_test = train_test_split(x_train_dummies, y, test_size=0.2, random_state=42)

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

print(accuracy_score(y_test, y_pred))
```
