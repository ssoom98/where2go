## 필요한 컬럼 뽑아서 합치기

```python
accomm_sel = accomm_con[['여행ID', '숙소유형코드', '결제금액_원']].rename(columns={'결제금액_원': '숙박결제금액'})
active_sel = active_con[['여행ID', '결제금액_원']].rename(columns={'결제금액_원': '활동결제금액'})
trans_sel = trans_con[['여행ID','결제금액_원']].rename(columns={'결제금액_원': '교통결제금액'})
pre_sel = pre_con[['여행ID', '결제금액_원']].rename(columns={'결제금액_원': '사전결제금액'})

select_merge = accomm_sel.merge(active_sel, on='여행ID', how='outer')
select_merge = select_merge.merge(trans_sel, on='여행ID', how='outer')
select_merge = select_merge.merge(pre_sel, on='여행ID', how='outer')
select_merge = select_merge.groupby('여행ID').agg({
    '숙소유형코드': 'first',             # 가장 앞에 있는 값
    '숙박결제금액': 'sum',
    '활동결제금액': 'sum',
    '교통결제금액': 'sum',
    '사전결제금액': 'sum'
})
select_merge['전체결제금액'] = select_merge['숙박결제금액'] + select_merge['활동결제금액'] + select_merge['교통결제금액'] + select_merge['사전결제금액']
select_merge = select_merge.dropna()

display(select_merge)
```

## 범주형 변수로 변경

```python
code_B = pd.read_csv(file_path + r'\tc_codeb_코드B.csv')

def dic_map(table, code, col):
    code_map = code_B[code_B['cd_a']== code][['cd_b', 'cd_nm']]
    code_map['cd_b'] = code_map['cd_b'].astype(int)
    code_dic = code_map.set_index('cd_b')['cd_nm'].to_dict()
    table[col] = table[col].astype("Int64").map(code_dic)
    return table

select_merge = dic_map(select_merge, 'HTY', '숙소유형코드')
display(select_merge)
```

## 독립표본 t-검정

```python
HotelUser = select_merge[select_merge['숙소유형코드']=='호텔']['전체결제금액']
MotelUser = select_merge[select_merge['숙소유형코드']=='모텔/여관']['전체결제금액']

from scipy.stats import ttest_ind

t_stat, p_val = ttest_ind(HotelUser, MotelUser)
print("p-value" , p_val)
```

p-value = 0.002 < 0.05 이므로 귀무가설을 기각하고 대립가설을 채택하게 됩니다.  
즉, 대림가설인 "모텔/여관 이용자는 호텔 이용자보다 전체 소비가 높지 않다" 를 채택하게 됩니다.
