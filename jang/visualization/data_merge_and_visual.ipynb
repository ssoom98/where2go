```python
import pandas as pd
import os

base_path = r'your path'

folder_path = [
    r'TS_csv',
    r'VS_csv'
]

file_names = os.listdir(os.path.join(base_path, folder_path[0]))

output_path = os.path.join(base_path, 'merged')
os.makedirs(output_path, exist_ok=True)

for file_name in file_names:
    df_list = []
    for folder in folder_path:
        file_path = os.path.join(base_path, folder, file_name)
        if os.path.exists(file_path):
            df = pd.read_csv(file_path)
            df_list.append(df)
        
    if df_list:
        merged_df = pd.concat(df_list, ignore_index=True)
        name, ext = os.path.splitext(file_name)
        new_file_name = f"{name}_East{ext}"
        merged_df.to_csv(os.path.join(output_path, new_file_name), index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 한글 및 마이너스 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 그림 크기 설정
plt.figure(figsize=(6, 4))

# 막대 색상 지정 (성별별 컬러 다르게 설정)
colors = {'남': '#4A90E2', '여': '#D0021B'}  # 성별에 따라 색상 지정
barplot = sns.countplot(data=df, x='성별', palette=colors)

# 제목 및 라벨
plt.title('성별 분포')
plt.xlabel('성별')
plt.ylabel('명 수')

# 배경선 제거
sns.despine()  # 테두리 제거
plt.grid(False)  # 배경 격자선 제거

# 막대 위에 인원 수 표시
for p in barplot.patches:
    count = int(p.get_height())
    barplot.annotate(f'{count}명',
                     xy=(p.get_x() + p.get_width() / 2, count),
                     ha='center', va='bottom',
                     fontsize=10)

plt.show()
```

## 여행 유형(가족끼리 여행, 1인 여행 등등)

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'  
plt.rcParams['axes.unicode_minus'] = False     

labels = [label for label in df['여행동행_유형'].value_counts().to_dict().keys()]

counts_dict = df['여행동행_유형'].value_counts().to_dict()
values = [counts_dict.get(label, 0) for label in labels] 

custom_colors = [
    'paleturquoise',  
    'palevioletred',  
    'palegreen',  
    'lightsalmon',  
    'lavender',  
    'orchid',   
    'lemonchiffon'  
]

fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(
    values, 
    labels=labels,
    colors=custom_colors,
    autopct="%1.1f%%",
    startangle=140, 
    wedgeprops=dict(width=0.7),  
    textprops=dict(color='black', fontsize=10)
)

centre_circle = plt.Circle((0,0), 0.3, fc='white') 
fig.gca().add_artist(centre_circle)

for autotext in autotexts:
    autotext.set_horizontalalignment('center')
    autotext.set_verticalalignment('center')

ax.axis('equal')
plt.title('여행 동행 유형 도넛 차트')
plt.tight_layout()
plt.show()
```

```python

companion = pd.read_csv(base_path + '\\tn_companion_info_동반자정보_B_East.csv')

com_id = {
    "COMPANION_SEQ": "동반자순번",
    "TRAVEL_ID": "여행ID",
    "REL_CD": "동반자관계코드",                  ## TCR
    "COMPANION_GENDER": "동반자성별",           ## GEN
    "COMPANION_AGE_GRP": "동반자연령대",        ## AGE
    "COMPANION_SITUATION": "동반자동반상황"     ## CST
}

companion.rename(columns=com_id, inplace=True)


file_path = r'C:\Users\hanaj\OneDrive\문서\[강남구혁신인재육성아카데미]데이터분석\미니프로젝트\국내여행로그데이터\동부권\TS_csv'
code_B = pd.read_csv(file_path + r'\tc_codeb_코드B.csv')

def dic_map(table, code, col):
    code_map = code_B[code_B['cd_a']== code][['cd_b', 'cd_nm']]
    code_map['cd_b'] = code_map['cd_b'].astype(int)
    code_dic = code_map.set_index('cd_b')['cd_nm'].to_dict()
    table[col] = table[col].astype("Int64").map(code_dic)
    return table
    

companion = dic_map(companion, 'TCR', '동반자관계코드')
companion = dic_map(companion, 'GEN', '동반자성별')
companion = dic_map(companion, 'AGE', '동반자연령대')
companion = dic_map(companion, 'CST', '동반자동반상황')

display(companion.head())
```

```python
df_filtered = df[df['여행동행_유형'] == '2인 여행(가족 외)']

merged = pd.merge(df_filtered, companion, on='여행ID', how='outer')

merged
# relation_counts = merged['동반자관계코드'].value_counts()
# relation_counts
```
