```python
import pandas as pd

# 파일 경로 설정
base_path = r"C:\Users\henaj\Downloads\278.국내 여행로그 데이터(동부권)\01-1.정식개방데이터\Training\01.원천데이터"
data_file = f"{base_path}\\tn_travel_여행_B.csv"
code_file = f"{base_path}\\tc_codeb_코드B.csv"
output_file = f"{base_path}\\tn_travel_여행_B_내용수정.csv"

# CSV 파일 불러오기
df_data = pd.read_csv(data_file, encoding='utf-8')
df_code = pd.read_csv(code_file, encoding='utf-8')

# 'MIS' 딕셔너리 생성: key 값을 문자열로 변환하고 공백 제거 (안정성 강화)
mis_dict = {str(row['cd_b']).strip(): row['cd_nm'] for _, row in df_code[df_code['cd_a'] == 'MIS'].iterrows()}

print("MIS 딕셔너리:", mis_dict) # 딕셔너리가 올바르게 생성되었는지 확인

# '여행목적', '개별미션', '미션우선도' 열 값 치환
# 데이터프레임의 열 값도 문자열로 변환하고 공백 제거 후, .get() 메서드 사용
# .get(key, default_value)는 키가 딕셔너리에 없으면 default_value를 반환하여 오류를 방지합니다.
df_data['여행 목적'] = df_data['여행 목적'].astype(str).str.strip().apply(lambda x: mis_dict.get(x, x))
df_data['개별미션'] = df_data['개별미션'].astype(str).str.strip().apply(lambda x: mis_dict.get(x, x))
df_data['미션우선도'] = df_data['미션우선도'].astype(str).str.strip().apply(lambda x: mis_dict.get(x, x))

# 변경된 데이터 저장
df_data.to_csv(output_file, index=False, encoding='utf-8-sig')

print("변환 완료!")
```

```python
import pandas as pd

# 파일 경로 설정
base_path = r"C:\Users\henaj\Downloads\278.국내 여행로그 데이터(동부권)\01-1.정식개방데이터\Training\01.원천데이터"
data_file = f"{base_path}\\tn_travel_여행_B_컬럼수정.csv"
code_file = f"{base_path}\\tc_codeb_코드B.csv"
output_file = f"{base_path}\\tn_travel_여행_B_내용수정.csv" # 출력 파일명 변경 (이전 파일 덮어쓰기 방지)

# CSV 파일 불러오기
df_data = pd.read_csv(data_file, encoding='utf-8')
df_code = pd.read_csv(code_file, encoding='utf-8')

# 'MIS' 딕셔너리 생성
# df_code의 cd_b 값을 문자열로 변환하고 양쪽 공백을 제거하여 딕셔너리 키로 사용합니다.
# 이는 데이터 불일치(타입, 공백) 문제를 방지합니다.
mis_dict = {str(row['cd_b']).strip(): row['cd_nm'] for _, row in df_code[df_code['cd_a'] == 'MIS'].iterrows()}

print("--- MIS 딕셔너리 확인 ---")
print(mis_dict)
print("-" * 30)

# 변환할 컬럼 리스트 (컬럼명이 이미 일치되었다는 전제하에 직접 명시)
# 만약 실제 컬럼명이 다르다면 이 리스트의 이름을 실제 컬럼명으로 수정해야 합니다.
columns_to_transform = ['여행목적', '개별미션', '미션우선도']

print("--- 컬럼 변환 시작 ---")
# 각 컬럼에 대해 변환 로직 적용
for col in columns_to_transform:
    # 해당 컬럼이 데이터프레임에 존재하는지 다시 한번 확인합니다.
    # 이 확인 절차는 KeyError 방지에 중요합니다.
    if col in df_data.columns:
        df_data[col] = df_data[col].astype(str).str.strip().apply(
            # 각 셀의 값(x)을 처리하는 람다 함수
            lambda x: ';'.join(
                [mis_dict.get(part.strip(), part.strip()) # 각 부분을 딕셔너리에서 찾거나, 없으면 원래 값 유지
                 for part in x.split(';')                  # ';' 기준으로 문자열 분할
                 if part.strip()]                           # 빈 문자열 제거
            ) if pd.notna(x) and ';' in x and x.strip()     # NaN이 아니고 ';'가 포함된 유효한 문자열일 경우
            else mis_dict.get(x, x)                        # ';'가 없거나 단일 값인 경우 바로 매핑
        )
        print(f"✅ '{col}' 컬럼 변환 완료!")
    else:
        print(f"❗ 경고: '{col}' 컬럼이 데이터프레임에 존재하지 않아 변환을 건너뛰었습니다. 컬럼명을 확인해주세요.")
print("-" * 30)

# 변경된 데이터 저장
df_data.to_csv(output_file, index=False, encoding='utf-8-sig')

print(f"✨ 모든 변환 완료! 결과는 '{output_file}'에 저장되었습니다.")

# 변환된 데이터의 일부를 확인하고 싶다면 아래 주석을 해제하세요.
# print("\n--- 변환 후 데이터 샘플 ---")
# print(df_data[columns_to_transform].head())
```

```python
import pandas as pd

# 파일 경로 설정
base_path = r"C:\Users\henaj\Downloads\278.국내 여행로그 데이터(동부권)\01-1.정식개방데이터\Training\01.원천데이터"
data_file = f"{base_path}\\tn_mvmn_consume_his_이동수단소비내역_B_컬럼수정.csv"
code_file = f"{base_path}\\tc_codeb_코드B.csv"
output_file = f"{base_path}\\tn_mvmn_consume_his_이동수단소비내역_B_내용수정10.csv" # 출력 파일명은 그대로 유지

# CSV 파일 불러오기
df_data = pd.read_csv(data_file, encoding='utf-8')
df_code = pd.read_csv(code_file, encoding='utf-8')

# 'MOV' 딕셔너리 생성: 이동수단구분용
# 딕셔너리 생성 시에도 key 값을 문자열로 변환하고 공백 제거를 명시적으로 적용
mov_dict = {str(row['cd_b']).strip(): row['cd_nm'] for _, row in df_code[df_code['cd_a'] == 'MOV'].iterrows()}

# 'PAY' 딕셔너리 생성: 결제방식구분용
# 딕셔너리 생성 시에도 key 값을 문자열로 변환하고 공백 제거를 명시적으로 적용 (핵심)
pay_dict = {str(row['cd_b']).strip(): row['cd_nm'] for _, row in df_code[df_code['cd_a'] == 'PAY'].iterrows()}

print("--- MOV 딕셔너리 확인 ---")
print(mov_dict)
print("-" * 30)
print("--- PAY 딕셔너리 확인 ---")
print(pay_dict)
print("-" * 30)

# '이동수단구분' 열 값 치환
# 데이터프레임의 열 값을 문자열로 변환하고 공백 제거 후, .get() 메서드 사용
# .get(key, default_value)는 키가 딕셔너리에 없으면 default_value를 반환하여 오류를 방지
df_data['이동수단구분'] = df_data['이동수단구분'].astype(str).str.strip().apply(lambda x: mov_dict.get(x, x))
print("✅ '이동수단구분' 컬럼 변환 완료!")

# '결제방식구분' 변환 (수정된 부분)
# 역시 .get() 메서드를 사용하여 키가 없을 때 원래 값을 유지하도록 변경
# 이 부분이 가장 중요합니다. 변환되지 않는 값이 있다면, 그 값이 pay_dict에 없다는 의미입니다.
df_data['결제방식구분'] = df_data['결제방식구분'].astype(str).str.strip().apply(lambda x: pay_dict.get(x, x))
print("✅ '결제방식구분' 컬럼 변환 완료!")

# 변경된 데이터 저장
df_data.to_csv(output_file, index=False, encoding='utf-8-sig')

print(f"✨ 모든 변환 완료! 결과는 '{output_file}'에 저장되었습니다.")

# 변환 후 '결제방식구분' 열의 고유 값을 확인하여 변환 여부 재확인
print("\n--- 변환 후 '결제방식구분' 고유 값 확인 ---")
print(df_data['결제방식구분'].unique())
```
