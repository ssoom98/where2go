```python
import pandas as pd
df = pd.read_csv("tn_activity_his_활동내역_B.csv", encoding = 'utf-8')
df.rename(columns = {
    'TRAVEL_ID':'여행ID',
    'VISIT_AREA_ID':'방문지ID',
    'ACTIVITY_TYPE_CD':'활동유형코드',
    'ACTIVITY_TYPE_SEQ':'활동유형순번',
    'ACTIVITY_ETC':'활동_기타',
    'ACTIVITY_DTL':'세부내역',
    'RSVT_YN':'예약여부',
    'EXPND_SE':'지출구분',
    'ADMISSION_SE':'입장료구분'
}, inplace=True)
```

```python
df['활동유형코드'] = df['활동유형코드'].replace({1:'취식', 2:'쇼핑/구매', 3:'체험활동/입장및관람', 4:'단순구경/산책/걷기', 5:'휴식', 6:'기타활동', 7:'환승/경유', 99: '없음'})
```

```python
df['지출구분'] = df['지출구분'].replace({1:'이 방문지에서 구매', 2:'이 방문지로 이동 중 구매', 3:'여행 출발 전 거주지 인근에서 미리 구매', 4:'비용 발생하지 않음'})
df['입장료구분'] = df['입장료구분'].replace({1:'체험', 2:'입장및관람'})
```

```python
# 1. 매핑 딕셔너리
mapping_지출구분 = {
    '1': '이 방문지에서 구매',
    '2': '이 방문지로 이동 중 구매',
    '3': '여행 출발 전 거주지 인근에서 미리 구매',
    '4': '비용 발생하지 않음'
    '5': '지출없음'
}

# 2. 매핑 함수 정의
def map_multiple_codes(cell):
    if pd.isna(cell):
        return cell
    codes = str(cell).split(';')
    mapped = [mapping_지출구분.get(code.strip(), code.strip()) for code in codes]
    return ';'.join(mapped)

# 3. 적용
df['지출구분'] = df['지출구분'].apply(map_multiple_codes)
```

```python
mapping_입장료 = {
    '1': '체험',
    '2': '입장및관람'
}

def map_multiple_codes2(cell):
    if pd.isna(cell):
        return cell
    codes = str(cell).split(';')
    mapped = [mapping_입장료.get(code.strip(), code.strip()) for code in codes]
    return ';'.join(mapped)

df['입장료구분'] = df['입장료구분'].apply(map_multiple_codes2)
```

```python
df.to_csv("tn_activity_his_활동내역_B_수정본.csv", index=False, encoding='utf-8-sig')
from google.colab import files
files.download("tn_activity_his_활동내역_B_수정본.csv")
```
